```python
transport_costs = {}
for index, row in transport_costs_df.iterrows():
  departure = row["GeoPoint#Name"]
  arrival = row["GeoPoint_1#Name"]
  cost = row["€/U"]
  if not departure in transport_costs:
    transport_costs[departure] = {}
  transport_costs[departure][arrival] = cost
```
